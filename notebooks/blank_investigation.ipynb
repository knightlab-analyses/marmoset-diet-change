```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
metadata = pd.read_csv('../data/blank_investigation/14577_20230502-100215.txt',sep='\t')
```

```python
prep = pd.read_csv('../data/blank_investigation/14577_prep_14581_20230404-143721.txt',sep='\t')
```

```python
prep.head()
```

```python
metadata[metadata['tube_id'].str.contains('BLANK')]
```

```python
metadata['sample_name'] = metadata['sample_name'].str.replace('14577.14577','14577.')
```

```python
meta_prep = pd.merge(metadata,prep,on='sample_name',how='left')
```

```python
plate_df = pd.read_csv('../data/blank_investigation/2023_05_02_AGMP_3XTG_P3_U19_1-3_Celeste_Marm_1_iSeq_pooling_df.txt', sep='\t', index_col=0)
plate_df
```

```python
plate_df[plate_df.Sample.str.contains("BLANK.Marmoset")]
```

```python
meta_prep_df = pd.merge(meta_prep, plate_df,left_on='tube_id',right_on='Sample', how='left')
```

```python
meta_prep_df
```

```python
meta_prep_df[meta_prep_df.sample_name.str.contains("BLANK")]
```

```python
meta_prep_df.loc[meta_prep_df['sample_name']=='14577.BLANK.Marmoset.4.12H']
```

```python
sns.boxplot(x='Blank',y='new_pooling_scaled',data=meta_prep_df)
```

```python
sns.boxplot(x='Blank',y='quality_filtered_reads',data=meta_prep_df,fliersize=0)
sns.stripplot(x='Blank',y='quality_filtered_reads',data=meta_prep_df,color='black',alpha=0.5)
```

```python
meta_prep_df['counts/pooled'] = meta_prep_df['quality_filtered_reads']/meta_prep_df['new_pooling_scaled']
```

```python
meta_prep_df['counts/pooled']
```

```python
sns.boxplot(x='Blank',y='counts/pooled',data=meta_prep_df.sort_values(by='counts/pooled',ascending=True),
            fliersize=0)
sns.stripplot(x='Blank',y='counts/pooled',data=meta_prep_df.sort_values(by='counts/pooled',ascending=True),
              color='black',alpha=0.5)
```

```python
sns.boxplot(x='Blank',y='counts/pooled',data=meta_prep_df.sort_values(by='counts/pooled',ascending=True),
            fliersize=0)
sns.stripplot(x='Blank',y='counts/pooled',data=meta_prep_df.sort_values(by='counts/pooled',ascending=True),
              color='black',alpha=0.5)
plt.yscale('log')
```

```python
ordered_cp_ratio = meta_prep_df.sort_values(by='counts/pooled',ascending=True)
ordered_cp_ratio.head(30)
#set threshold at 3000 quality_filtered_reads per new_pooling_scaled
```
